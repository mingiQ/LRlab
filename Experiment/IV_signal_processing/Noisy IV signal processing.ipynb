## 1. `scipy.signal` method

```python
import numpy as np
import os
import matplotlib.pyplot as plt
import sys
import pandas as pd
import seaborn as sns
sns.set_theme(style="whitegrid")
```

```python
from scipy.fft import fft, fftfreq, fftshift
from scipy.signal import lfilter, decimate, filtfilt, iirfilter
```

```python
wdir = './I-V test for dV-dI/'

ivdata = os.listdir(wdir)

iv1000 = [x for x in ivdata if '1000pts' in x]                 # 1000 pts
iv5000 = [x for x in ivdata if '5000pts' in x]                 # 5000 pts
iv10000 = [x for x in ivdata if '10000pts' in x]               # 10000 pts
iv_fil = [x for x in ivdata if 'filtered' in x]                # Lockin measurement
```

```python
ivdata
```

### FFT of one of the IV data

```python
iv = iv10000[-1]

sample = np.loadtxt(wdir+iv, delimiter=',')
I = sample[:,0]
V = sample[:,1]
Vf = fft(V)
If = fftfreq(len(I), np.abs(I[1]-I[2]))

Vfs = fftshift(Vf)
Ifs = fftshift(If)
plt.plot(Ifs, Vfs,'.')
plt.yscale('log')
```

mostly in 0 frequency...?

### Filter types ;  IIR : Infinite Impulse Response,  FIR : finite Impulse Response 
--> we may need infinite impulse response filter

```python
# In[tesing IIR --> decimation]

iv = iv5000[3]

sample = np.loadtxt(wdir+iv, delimiter=',')
I_raw = sample[:,0]
V_raw = sample[:,1]
fs = 30

b, a = iirfilter(4, Wn=2.5, fs=fs, btype="low", ftype="butter")
print(b, a, sep="\n")

V_iir = filtfilt(b,a,V_raw)

downsampling_factor = int(len(V_iir)/50)

Vdem = decimate(V_iir, downsampling_factor)
Inew = np.linspace(I[0], I[-1], len(Vdem))

dvdi = (Vdem[1:]-Vdem[:-1])/(Inew[1:]-Inew[:-1])
ilist = 0.5*(Inew[1:]+Inew[:-1])

#plt.figure(figsize=(5,10))
plt.subplot(211)
plt.plot(Inew, Vdem, '.')
plt.plot(I_raw, V_iir, 'g-', alpha=0.4)
plt.plot(I_raw, V_raw, 'r-', alpha=0.4)

plt.subplot(212)
plt.plot(ilist, dvdi, '.-')
```

right hand part of the IV shows phase lag and decreased signal

Just using decimation,

```python
# In[ decimation]

for i in range(6):
    iv = iv10000[i]
    
    
    sample = np.loadtxt(wdir+iv, delimiter=',')
    I_raw = sample[:,0]
    V_raw = sample[:,1]
    
    V_iir = V_raw
    downsampling_factor = int(len(V_iir)/50)
    
    Vdem = decimate(x=V_iir, q=downsampling_factor, n=3, ftype='iir' ,zero_phase=True)           
    # x : data will be filtered ,  q ; downsampling filter --> need to find optimal value, n: order of the filter (seems n=3 is optimal..?)  
    #ftype: filter type, i wanted to use IIR filter, zero_phase : zeroing the phase lag
    Inew = np.linspace(I[0], I[-1], len(Vdem))
    
    dvdi = (Vdem[1:]-Vdem[:-1])/(Inew[1:]-Inew[:-1])
    ilist = 0.5*(Inew[1:]+Inew[:-1])
    
    ref1 = np.loadtxt(wdir+iv_fil[0], skiprows=2, delimiter=',')
    ref2 = np.loadtxt(wdir+iv_fil[1], skiprows=2, delimiter=',')
    ref3 = np.loadtxt(wdir+iv_fil[2], skiprows=2, delimiter=',')
    
    dvdi_ref1 = ref1[:,1]
    dvdi_ref2 = ref2[:,1]
    dvdi_ref3 = ref3[:,1]
    
    i_ref = ref1[:,0]
    
    plt.figure(figsize=(5,10))
    plt.subplot(211)
    plt.plot(Inew, Vdem, '.', label=iv)
    plt.plot(I_raw, V_iir, 'g-', alpha=0.4)
    plt.plot(I_raw, V_raw, 'r-', alpha=0.4)
    plt.legend()
    
    plt.subplot(212)
    plt.plot(ilist, dvdi*1e5, '.-')
    plt.plot(i_ref, dvdi_ref1, 'g-', alpha=0.4)
    plt.plot(i_ref, dvdi_ref2, 'r-', alpha=0.4)
    plt.plot(i_ref, dvdi_ref3, 'y-', alpha=0.4)
```
